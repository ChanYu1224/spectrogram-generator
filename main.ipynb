```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
ja_data, ja_sr = librosa.load('./audio/konichiha.wav')
en_data, en_sr = librosa.load('./audio/hello.wav')
```

```python
plt.plot(ja_data, color='black')
plt.show()
```

```python
plt.plot(en_data, color='black')
plt.show()
```

```python
ja_spec = np.abs(librosa.core.stft(ja_data))
en_spec = np.abs(librosa.core.stft(en_data))
```

```python
fig, ax = plt.subplots()
img = librosa.display.specshow(librosa.amplitude_to_db(ja_spec,ref=np.max),y_axis='log', ax=ax)
```

```python
fig, ax = plt.subplots()
img = librosa.display.specshow(librosa.amplitude_to_db(en_spec,ref=np.max),y_axis='log', ax=ax)
```
